# 준비

## 라이브러리

```python
import scipy
import numpy as np
from matplotlib import pyplot as plt

import warnings
warnings.filterwarnings(action='ignore')

import tensorflow as tf; print(tf.__version__);
from tensorflow import keras
from tensorflow.keras import backend as K
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications import inception_v3
```

## 세팅

```python
# 모델을 훈련하지 않습니다. 이 명령은 모든 훈련 연산을 비활성화합니다
K.set_learning_phase(0)

# 합성곱 기반층만 사용한 인셉션 V3 네트워크를 만듭니다. 사전 훈련된 ImageNet 가중치와 함께 모델을 로드합니다
model = inception_v3.InceptionV3(weights='imagenet',include_top=False)
```

# 모델링

## 레이어 contributions

```python
# 최대화하려는 손실에 층의 활성화가 기여할 양을 정합니다.
# model.summary()를 사용하면 모든 층 이름을 확인할 수 있습니다
layer_contributions = {
    'mixed2': 0.2,
    'mixed3': 3.,
    'mixed4': 2.,
    'mixed5': 1.5,
}
```

## 손실 정의
* 손실 텐서를 정의. 위에서 선택한 층의 활성화에 대한 L2 노름의 가중치 합

```python
# 층 이름과 층 객체를 매핑한 딕셔너리
layer_dict = dict([(layer.name, layer) for layer in model.layers])

# 손실을 정의, 각 층의 기여분을 이 스칼라 변수에 추가
loss = K.variable(0.)
for layer_name in layer_contributions:
    coeff = layer_contributions[layer_name]
    
    # 층의 출력
    activation = layer_dict[layer_name].output

    scaling = K.prod(K.cast(K.shape(activation), 'float32'))
    # 층 특성의 L2 노름의 제곱을 손실에 추가합니다. 이미지 테두리는 제외하고 손실에 추가합니다.
    loss = loss + coeff * K.sum(K.square(activation[:, 2: -2, 2: -2, :])) / scaling
```

## loss, grad 계산

```python
# 생성된 딥드림 이미지를 저장
dream = model.input

# 손실 그래디언트를 계산
grads = K.gradients(loss, dream)[0]

# 그래디언트를 정규화 (중요함)
grads /= K.maximum(K.mean(K.abs(grads)), 1e-7)

# loss, grad 를 계산할 함수
outputs = [loss, grads]
fetch_loss_and_grads = K.function([dream], outputs) # input=[dream], output=outputs

def eval_loss_and_grads(x):
    outs = fetch_loss_and_grads([x])
    loss_value = outs[0] # loss
    grad_values = outs[1] # grads
    return loss_value, grad_values

def gradient_ascent(x, iterations, step, max_loss=None):
    for i in range(iterations):
        loss_value, grad_values = eval_loss_and_grads(x)
        if max_loss is not None and loss_value > max_loss:
            break
        print('...', i, '번째 손실 :', loss_value)
        x += step * grad_values
    return x
```

## 딥드림

먼저 이미지를 처리하기 위한 스케일(옥타브)리스트 정의. 스케일은 이전 스케일보다 1.4배 크다.

![deep dream process](https://s3.amazonaws.com/book.keras.io/img/ch8/deepdream_process.png)

```python
# 이미지 조절
def resize_img(img, size):
    img = np.copy(img)
    factors = (1, float(size[0]) / img.shape[1], float(size[1]) / img.shape[2], 1)
    return scipy.ndimage.zoom(img, factors, order=1)

# 인셉션V3 전처리 
def preprocess_image(image_path):
    img = image.load_img(image_path)
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = inception_v3.preprocess_input(img)
    return img

# 전처리 복원
def deprocess_image(x):
    # 넘파이 배열을 적절한 이미지 포맷으로 변환하는 유틸리티 함수
    if K.image_data_format() == 'channels_first':
        x = x.reshape((3, x.shape[2], x.shape[3]))
        x = x.transpose((1, 2, 0))
    else: x = x.reshape((x.shape[1], x.shape[2], 3))
    x /= 2.
    x += 0.5
    x *= 255.
    x = np.clip(x, 0, 255).astype('uint8')
    return x

# 이미지 저장
def save_img(img, fname):
    pil_img = deprocess_image(np.copy(img))
    image.save_img(fname, pil_img)
```

```python
step = 0.01  # 경상 상승법 단계 크기
num_octave = 3  # 경사 상승법을 실행할 스케일 단계 횟수
octave_scale = 1.4  # 스케일 간의 크기 비율
iterations = 20  # 스케일 단계마다 수행할 경사 상승법 횟수

# 손실이 10보다 커지면 그림이 너무 이상해지므로 중단
max_loss = 10.

# 사용할 이미지 경로
base_image_path = './Image_src/original_photo_deep_dream.jpg'

# 기본 이미지를 넘파이 배열로
img = preprocess_image(base_image_path)

# 튜플의 리스트를 준비
original_shape = img.shape[1:3]
successive_shapes = [original_shape]
for i in range(1, num_octave):
    shape = tuple([int(dim / (octave_scale ** i)) for dim in original_shape])
    successive_shapes.append(shape)

# 이 리스트를 크기 순으로
successive_shapes = successive_shapes[::-1]

# 이미지의 넘파이 배열을 가장 작은 스케일로 변경
original_img = np.copy(img)
shrunk_original_img = resize_img(img, successive_shapes[0])

for shape in successive_shapes:
    print('처리할 이미지 크기', shape)
    img = resize_img(img, shape)
    img = gradient_ascent(img,
                          iterations=iterations,
                          step=step,
                          max_loss=max_loss)
    upscaled_shrunk_original_img = resize_img(shrunk_original_img, shape)
    same_size_original = resize_img(original_img, shape)
    lost_detail = same_size_original - upscaled_shrunk_original_img

    img += lost_detail
    shrunk_original_img = resize_img(original_img, shape)
    save_img(img, fname='./tmp/dream_at_scale_' + str(shape) + '.png')

save_img(img, fname='./tmp/final_dream.png')
```
